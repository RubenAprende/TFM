```python
from GoogleNews import GoogleNews
from newspaper import Config
from newspaper import Article
import pandas as pd
```

```python
googlenews=GoogleNews(start='01/01/2021',end='01/31/2020')
googlenews.search('AMZN')
result=googlenews.result()
df=pd.DataFrame(result)
df.head()
```

```python
for i in range(2,20):
    googlenews.getpage(i)
    result=googlenews.result()
    df=pd.DataFrame(result)
df
```

```python
config = Config()
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36"
config.browser_user_agent = user_agent
config.request_timeout = 10

mylist = list()
for ind in df.index:
    dict={}
    article = Article(df['link'][ind])
    article.download()
    try:
        article.parse()
        article.nlp()
        dict['Date']=df['date'][ind]
        dict['Media']=df['media'][ind]
        dict['Title']=article.title
        dict['Article']=article.text
        dict['Summary']=article.summary
        mylist.append(dict)
    except:
        print("Pass: ", df['link'][ind])
        pass
news_df=pd.DataFrame(mylist)
news_df
```

```python
from GoogleNews import GoogleNews
from newspaper import Article
from newspaper import Config
import pandas as pd
import nltk
#config will allow us to access the specified url for which we are #not authorized. Sometimes we may get 403 client error while parsing #the link to download the article.
nltk.download('punkt')

googlenews=GoogleNews(start='05/01/2020',end='05/31/2020')
googlenews.search('AMZN')
result=googlenews.result()
df=pd.DataFrame(result)
print(df.head())
for i in range(2,20):
    googlenews.getpage(i)
    result=googlenews.result()
    df=pd.DataFrame(result)
list=[]
for ind in df.index:
    dict={}
    article = Article(df['link'][ind])
    article.download()
    article.parse()
    article.nlp()
    dict['Date']=df['date'][ind]
    dict['Media']=df['media'][ind]
    dict['Title']=article.title
    dict['Article']=article.text
    dict['Summary']=article.summary
    list.append(dict)
news_df=pd.DataFrame(list)
news_df
```
